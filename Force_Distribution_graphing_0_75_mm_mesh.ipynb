```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as pt
```

```python
forces = pd.read_excel("Layer_forces_BCC.xlsx")
```

```python
displacement = list(map(lambda x: -x, list(forces["Displacement"])))
print(displacement)
```

```python
style = ["k-", "g-", "r-", "b-", "m-", "c-"]
layer_forces = {}
#pt.xkcd()
for i in range(1, 6):
    layer_forces["force" + str(i)] = list(forces["Layer " + str(i)])
    pt.plot(displacement, layer_forces["force" + str(i)], style[i - 1], label = ("Layer " + str(i)))

pt.xlabel("Displacement [mm]")
pt.ylabel("Force [N]")
pt.grid(True)
pt.legend()
pt.show
pt.savefig('Layer_Force_Distrib_normal.png', dpi=300, bbox_inches='tight')
```
